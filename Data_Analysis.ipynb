```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
# Average Running time
z = [856.5512, 185.5181, 262.9264, 243.5685, 138.7887, 105.4926, 26.0647, 25.9631, 27.3209, 24.0474, 13.8491, 13.9613, 0.6927, 0.2021, 0.2375, 0.2068]
# Tag
x = [15007, 2349, 3960, 3385, 1555, 1141, 216, 216, 216, 464, 193, 193, 12, 9, 9, 9]
# Length
y = [2955127, 973603, 973472, 939956, 603302, 397198, 110856, 110856, 110856, 93236, 58711, 58006, 2142, 346, 256, 251]
```

```python
# Plot  
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, 'b.')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

# Test Qt
qt = 16

# Solver
A = np.ones((qt, 3))
b = np.zeros((qt, 1))
for j in range(qt):
    A[j, 0] = x[j]
    A[j, 1] = y[j]
    b[j, 0] = z[j]

# X=(AT*A)^(-1)*AT*b
A_T = A.T
A1 = np.dot(A_T, A)
A2 = np.linalg.inv(A1)
A3 = np.dot(A2, A_T)
X = np.dot(A3, b)
print('Regression：z = %.3f * x + %.3f * y + %.3f' % (X[0, 0], X[1, 0], X[2, 0]))
```
